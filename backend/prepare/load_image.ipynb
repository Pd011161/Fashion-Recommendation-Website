```python
import os
import requests
import gdown

def save_img_links(df, save_dir):
    os.makedirs(save_dir, exist_ok=True)
    for idx, row in df.iterrows():
        url = str(row['URL'])
        img_id = str(row['id'])
        file_path = os.path.join(save_dir, f"{img_id}.jpg")
        
        try:
            # ถ้าเป็น Google Drive link ให้ใช้ gdown
            if 'drive.google.com' in url or 'drive.usercontent.google.com' in url:
                # หา file id
                if "id=" in url:
                    file_id = url.split("id=")[-1].split("&")[0]
                else:
                    # บางกรณีลิงก์รูปแบบอื่น
                    file_id = url.split('/')[-2]
                gdown_url = f"https://drive.google.com/uc?id={file_id}"
                gdown.download(gdown_url, file_path, quiet=False)
                print(f"Saved (gdown): {file_path}")
            else:
                # ลิงก์ภาพปกติ ใช้ requests
                response = requests.get(url, timeout=15)
                response.raise_for_status()
                with open(file_path, 'wb') as f:
                    f.write(response.content)
                print(f"Saved: {file_path}")
        except Exception as e:
            print(f"Error saving {img_id}: {e}")
```

```python
shirt_save_dir = '/Users/Aksorn_AI/Desktop/RECOMENTATION_SYSTEM/backend/datalake/shirt'
save_img_link(df_shirt, shirt_save_dir)
```

```python
pants_save_dir = '/Users/Aksorn_AI/Desktop/RECOMENTATION_SYSTEM/backend/datalake/pants'
save_img_link(df_pants, pants_save_dir)
```
