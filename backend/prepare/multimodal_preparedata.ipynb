````python
import pandas as pd

def clean_img_url_column(df, url_col="IMG_URL"):
    def convert_drive_url(url):
        if isinstance(url, str) and "drive.google.com" in url:
            if "id=" in url:
                file_id = url.split("id=")[-1].split("&")[0]
            elif "/d/" in url:
                file_id = url.split("/d/")[1].split("/")[0]
            else:
                return url  # Not convertible
            return f"https://drive.google.com/uc?export=download&id={file_id}"
        return url

    df[url_col] = df[url_col].apply(convert_drive_url)
    return df

df = clean_img_url_column(df)      # ‡πÅ‡∏õ‡∏•‡∏á IMG_URL ‡∏ó‡∏±‡πâ‡∏á‡∏´‡∏°‡∏î‡πÉ‡∏´‡πâ‡πÉ‡∏ä‡πâ‡πÑ‡∏î‡πâ
# ‡∏ó‡∏î‡∏™‡∏≠‡∏ö‡∏î‡∏π‡∏ú‡∏•
print(df[["Name", "IMG_URL"]].head())
````

````python
import pandas as pd
import os, json, time
from dotenv import load_dotenv
from langchain_openai import ChatOpenAI
from langchain_core.messages import HumanMessage

# ===== Load API Key =====
load_dotenv()
openapi_key = os.getenv("OPENAI_API_KEY")

# ===== Initialize LLM =====
llm = ChatOpenAI(
    model="gpt-4o",
    temperature=0.4,
    max_tokens=2048,
    api_key=openapi_key
)

# ===== Prompt Setup =====
system_prompt = (
    "You are a fashion product expert who writes bilingual (Thai and English) "
    "product descriptions and metadata JSON for pants."
)

user_prompt_text = """Given this pants image (you may ignore the person wearing it):

1. Write a detailed **Thai product description** for online selling. The Thai description must include:
- ‡∏ß‡∏±‡∏™‡∏î‡∏∏
- ‡∏™‡∏µ
- ‡∏ó‡∏£‡∏á/‡∏Ñ‡∏ß‡∏≤‡∏°‡∏û‡∏≠‡∏î‡∏µ
- ‡πÄ‡∏≠‡∏ß
- ‡∏î‡∏µ‡πÄ‡∏ó‡∏•
- ‡∏™‡πÑ‡∏ï‡∏•‡πå
- ‡πÇ‡∏≠‡∏Å‡∏≤‡∏™‡∏ó‡∏µ‡πà‡πÄ‡∏´‡∏°‡∏≤‡∏∞‡∏™‡∏°‡πÉ‡∏ô‡∏Å‡∏≤‡∏£‡∏™‡∏ß‡∏°‡πÉ‡∏™‡πà
- ‡∏Ñ‡∏≥‡πÅ‡∏ô‡∏∞‡∏ô‡∏≥‡πÉ‡∏ô‡∏Å‡∏≤‡∏£‡∏î‡∏π‡πÅ‡∏•‡∏£‡∏±‡∏Å‡∏©‡∏≤

Thai description example:

‡∏Å‡∏≤‡∏á‡πÄ‡∏Å‡∏á‡∏ó‡∏£‡∏á‡∏´‡∏•‡∏ß‡∏°‡∏Ç‡∏≤‡∏Å‡∏£‡∏∞‡∏ö‡∏≠‡∏Å ‡∏™‡∏µ‡πÄ‡∏Ç‡∏µ‡∏¢‡∏ß‡πÄ‡∏Ç‡πâ‡∏° ‡∏™‡πÑ‡∏ï‡∏•‡πå‡∏°‡∏¥‡∏ô‡∏¥‡∏°‡∏≠‡∏• ‡∏à‡∏≤‡∏Å KING STUDIO

‡∏Å‡∏≤‡∏á‡πÄ‡∏Å‡∏á‡∏Ç‡∏≤‡∏¢‡∏≤‡∏ß‡∏ó‡∏£‡∏á‡∏´‡∏•‡∏ß‡∏°‡∏ó‡∏µ‡πà‡∏≠‡∏≠‡∏Å‡πÅ‡∏ö‡∏ö‡∏°‡∏≤‡πÄ‡∏û‡∏∑‡πà‡∏≠‡∏Ñ‡∏ß‡∏≤‡∏°‡πÄ‡∏£‡∏µ‡∏¢‡∏ö‡∏á‡πà‡∏≤‡∏¢‡πÅ‡∏•‡∏∞‡∏Ñ‡∏ß‡∏≤‡∏°‡∏™‡∏ö‡∏≤‡∏¢‡πÉ‡∏ô‡∏Å‡∏≤‡∏£‡∏™‡∏ß‡∏°‡πÉ‡∏™‡πà ‡∏î‡πâ‡∏ß‡∏¢‡πÇ‡∏ó‡∏ô‡∏™‡∏µ‡πÄ‡∏Ç‡∏µ‡∏¢‡∏ß‡πÄ‡∏Ç‡πâ‡∏°‡∏ó‡∏µ‡πà‡∏Ñ‡∏•‡∏≤‡∏™‡∏™‡∏¥‡∏Å ‡∏™‡∏≤‡∏°‡∏≤‡∏£‡∏ñ‡∏à‡∏±‡∏ö‡∏Ñ‡∏π‡πà‡∏Å‡∏±‡∏ö‡πÄ‡∏™‡∏∑‡πâ‡∏≠‡πÑ‡∏î‡πâ‡∏´‡∏•‡∏≤‡∏Å‡∏´‡∏•‡∏≤‡∏¢‡∏™‡πÑ‡∏ï‡∏•‡πå ‡∏ó‡∏±‡πâ‡∏á‡πÅ‡∏ö‡∏ö‡∏ó‡∏≤‡∏á‡∏Å‡∏≤‡∏£‡πÅ‡∏•‡∏∞‡∏•‡∏≥‡∏•‡∏≠‡∏á ‡∏ï‡∏±‡∏ß‡∏Å‡∏≤‡∏á‡πÄ‡∏Å‡∏á‡∏°‡∏µ‡∏ó‡∏£‡∏á‡∏Å‡∏£‡∏∞‡∏ö‡∏≠‡∏Å‡∏ï‡∏£‡∏á ‡∏ä‡πà‡∏ß‡∏¢‡πÄ‡∏û‡∏¥‡πà‡∏°‡∏Ñ‡∏ß‡∏≤‡∏°‡πÇ‡∏õ‡∏£‡πà‡∏á‡∏™‡∏ö‡∏≤‡∏¢‡∏Ç‡∏ì‡∏∞‡∏™‡∏ß‡∏°‡πÉ‡∏™‡πà ‡πÅ‡∏•‡∏∞‡∏î‡∏µ‡πÄ‡∏ó‡∏•‡∏Å‡∏£‡∏∞‡πÄ‡∏õ‡πã‡∏≤‡∏Ç‡πâ‡∏≤‡∏á‡πÄ‡∏û‡∏∑‡πà‡∏≠‡∏Å‡∏≤‡∏£‡πÉ‡∏ä‡πâ‡∏á‡∏≤‡∏ô‡∏ó‡∏µ‡πà‡∏™‡∏∞‡∏î‡∏ß‡∏Å

‡∏ß‡∏±‡∏™‡∏î‡∏∏: ‡∏ú‡πâ‡∏≤‡πÇ‡∏û‡∏•‡∏µ‡πÄ‡∏≠‡∏™‡πÄ‡∏ï‡∏≠‡∏£‡πå‡∏ú‡∏™‡∏°‡∏Ñ‡∏≠‡∏ï‡∏ï‡∏≠‡∏ô ‡πÉ‡∏´‡πâ‡∏™‡∏±‡∏°‡∏ú‡∏±‡∏™‡πÄ‡∏ö‡∏≤‡πÅ‡∏•‡∏∞‡πÑ‡∏°‡πà‡∏¢‡∏±‡∏ö‡∏á‡πà‡∏≤‡∏¢

‡∏™‡∏µ: ‡πÄ‡∏Ç‡∏µ‡∏¢‡∏ß‡πÄ‡∏Ç‡πâ‡∏° (Dark Olive)

‡∏ó‡∏£‡∏á: ‡∏ó‡∏£‡∏á‡∏´‡∏•‡∏ß‡∏° (Relaxed Fit) ‡∏Ç‡∏≤‡∏Å‡∏£‡∏∞‡∏ö‡∏≠‡∏Å‡∏ï‡∏£‡∏á

‡πÄ‡∏≠‡∏ß: ‡πÄ‡∏≠‡∏ß‡∏¢‡∏≤‡∏á‡∏¢‡∏∑‡∏î‡∏î‡πâ‡∏≤‡∏ô‡∏´‡∏•‡∏±‡∏á ‡πÄ‡∏û‡∏¥‡πà‡∏°‡∏Ñ‡∏ß‡∏≤‡∏°‡∏¢‡∏∑‡∏î‡∏´‡∏¢‡∏∏‡πà‡∏ô

‡∏î‡∏µ‡πÄ‡∏ó‡∏•: ‡∏Å‡∏£‡∏∞‡πÄ‡∏õ‡πã‡∏≤‡∏Ç‡πâ‡∏≤‡∏á 2 ‡∏Ç‡πâ‡∏≤‡∏á / ‡∏Å‡∏£‡∏∞‡πÄ‡∏õ‡πã‡∏≤‡∏´‡∏•‡∏±‡∏á 1 ‡πÉ‡∏ö

‡∏™‡πÑ‡∏ï‡∏•‡πå: ‡∏°‡∏¥‡∏ô‡∏¥‡∏°‡∏≠‡∏• ‡πÄ‡∏£‡∏µ‡∏¢‡∏ö‡∏á‡πà‡∏≤‡∏¢ ‡πÉ‡∏™‡πà‡∏á‡πà‡∏≤‡∏¢‡∏ó‡∏∏‡∏Å‡∏ß‡∏±‡∏ô

‡πÄ‡∏´‡∏°‡∏≤‡∏∞‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö: ‡πÉ‡∏™‡πà‡πÄ‡∏ó‡∏µ‡πà‡∏¢‡∏ß ‡πÄ‡∏î‡∏¥‡∏ô‡πÄ‡∏•‡πà‡∏ô ‡∏ó‡∏≥‡∏á‡∏≤‡∏ô ‡∏´‡∏£‡∏∑‡∏≠‡∏™‡πÑ‡∏ï‡∏•‡πå Smart Casual

‡∏Ñ‡∏≥‡πÅ‡∏ô‡∏∞‡∏ô‡∏≥‡πÉ‡∏ô‡∏Å‡∏≤‡∏£‡∏î‡∏π‡πÅ‡∏•: ‡∏ã‡∏±‡∏Å‡πÄ‡∏Ñ‡∏£‡∏∑‡πà‡∏≠‡∏á‡πÉ‡∏ô‡∏ô‡πâ‡∏≥‡πÄ‡∏¢‡πá‡∏ô ‡∏´‡∏•‡∏µ‡∏Å‡πÄ‡∏•‡∏µ‡πà‡∏¢‡∏á‡∏Å‡∏≤‡∏£‡πÉ‡∏ä‡πâ‡∏™‡∏≤‡∏£‡∏ü‡∏≠‡∏Å‡∏Ç‡∏≤‡∏ß ‡πÅ‡∏•‡∏∞‡∏Ñ‡∏ß‡∏£‡∏£‡∏µ‡∏î‡∏î‡πâ‡∏ß‡∏¢‡πÑ‡∏ü‡∏≠‡πà‡∏≠‡∏ô


2. Write a detailed **English product description** with the same structure. It must include:
- Material
- Color
- Fit
- Waist
- Details
- Style
- Ideal occasions
- Care instructions

English description example:

KING STUDIO Minimalist Wide-Leg Trousers ‚Äì Dark Olive

Elevate your everyday look with these relaxed-fit, wide-leg trousers from KING STUDIO. Designed with a timeless minimalist aesthetic, the dark olive color adds a touch of sophistication while maintaining versatility. The trousers feature a comfortable straight-leg cut and practical side pockets, making them ideal for both casual and semi-formal settings.

Material: Polyester-cotton blend ‚Äì lightweight and wrinkle-resistant

Color: Dark olive

Fit: Relaxed fit with straight wide legs

Waist: Elastic waistband at the back for added comfort

Details: Two side pockets / One back pocket

Style: Minimalist, easy-to-match, and everyday-friendly

Ideal for: Casual outings, city walks, office wear, or smart casual looks

Care Instructions: Machine wash cold. Do not bleach. Iron on low heat if necessary.


3. Generate metadata in JSON format using English fields, suitable for storing in a database.

Your metadata JSON should have the following structure and keys:

{
  "product_name": (Full product name with brand and key design detail),
  "brand": (Brand name),
  "category": "Men's Clothing",
  "type": "Pants",
  "pants_type": (e.g., "Trousers", "Chinos", "Jeans", "Joggers", "Cargo Pants"),
  "style": (e.g., "Minimalist", "Streetwear", "Smart Casual", "Formal"),
  "color": (e.g., "Dark Olive", "Beige", "Black with pinstripes"),
  "material": (e.g., "Polyester-cotton blend", "Denim", "Linen"),
  "features": {
    "fit": (e.g., "Slim fit", "Regular fit", "Relaxed fit"),
    "leg_style": (e.g., "Straight", "Tapered", "Wide-leg", "Cropped"),
    "waist": (e.g., "Elastic waistband", "Button and zip closure", "Drawstring"),
    "pockets": {
      "side": (number of side pockets),
      "back": (number of back pockets)
    },
    "wrinkle_resistant": true or false,
    "lightweight": true or false
  },
  "ideal_for": [
    (e.g., "Office wear", "City walks", "Casual outings", "Smart casual occasions")
  ],
  "care_instructions": [
    (e.g., "Machine wash cold", "Do not bleach", "Iron on low heat if necessary")
  ],
  "description": (Concise product summary for catalog listing)
}

Respond in this exact format:
---
[TH]
(Thai description)
[EN]
(English description)
[METADATA]
```json
(your metadata JSON)
```"""

# ===== Send Image + Prompt =====
def generate_description_with_langchain(image_url: str) -> str:
    human_message = HumanMessage(
        content=[
            {"type": "text", "text": user_prompt_text},
            {"type": "image_url", "image_url": {"url": image_url}},
        ]
    )
    response = llm.invoke([{"role": "system", "content": system_prompt}, human_message])
    print("Generate:\n", response.content)
    return response.content
````

````python
import requests
import json
import time
import pandas as pd

# ===== ‡πÅ‡∏õ‡∏•‡∏á Google Drive URL ‡πÉ‡∏´‡πâ‡πÉ‡∏ä‡πâ‡πÑ‡∏î‡πâ‡∏Å‡∏±‡∏ö requests/GPT-4V =====

# ===== ‡∏•‡∏π‡∏õ‡∏™‡∏£‡πâ‡∏≤‡∏á Description =====
def loop_gen_desc(df):
    for idx, row in df.iterrows():
        en_desc = row["EN_Description"]

        if pd.isna(en_desc) or en_desc.strip() == "":
            # ‡πÅ‡∏õ‡∏•‡∏á‡∏Ñ‡∏≠‡∏•‡∏±‡∏°‡∏ô‡πå‡∏´‡∏•‡∏±‡∏á‡∏à‡∏≤‡∏Å‡∏ó‡∏µ‡πà‡πÄ‡∏£‡∏≤‡πÉ‡∏ä‡πâ‡∏ï‡∏£‡∏ß‡∏à‡πÄ‡∏™‡∏£‡πá‡∏à‡πÅ‡∏•‡πâ‡∏ß
            for col in ["Name", "TH_Description", "EN_Description", "metadata"]:
                df.at[idx, col] = str(df.at[idx, col])

            raw_url = row["IMG_URL"]
            url = raw_url

            print(f"Processing index {idx}...")
            print("üîó Checking URL:", url)

            try:
                response = requests.get(url, timeout=10)
                if response.status_code != 200 or "html" in response.headers.get("Content-Type", ""):
                    print(f"‚ùå URL not accessible or not an image (status code: {response.status_code})")
                    continue

                # ===== Generate Description =====
                result = generate_description_with_langchain(url)

                if '[TH]' in result and '[EN]' in result and '[METADATA]' in result:
                    th_desc = result.split('[TH]')[1].split('[EN]')[0].strip()
                    en_desc = result.split('[EN]')[1].split('[METADATA]')[0].strip()
                    metadata_str = result.split('```json')[1].split('```')[0].strip()
                    metadata = json.loads(metadata_str)

                    product_name = metadata.get("product_name", "Unnamed Product")
                    df.at[idx, "Name"] = product_name
                    df.at[idx, "TH_Description"] = th_desc
                    df.at[idx, "EN_Description"] = en_desc
                    df.at[idx, "metadata"] = json.dumps(metadata, ensure_ascii=False)

                    print("‚úÖ Name:", product_name)
                    print("TH_Description:\n", th_desc)
                    print("EN_Description:\n", en_desc)
                    print("Metadata:\n", metadata)
                    print("\n")
                    print("-" * 80)
                    print("\n")
                else:
                    raise ValueError("Missing one of [TH], [EN], or [METADATA] in response.")

            except Exception as e:
                print(f"‚ùå Failed at index {idx}: {e}")
                continue

            time.sleep(1.5)

    return df
````
