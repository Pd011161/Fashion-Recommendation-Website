Clean link 

```python
import pandas as pd

def clean_img_url_column(df, url_col="URL"):
    def convert_drive_url(url):
        if isinstance(url, str) and "drive.google.com" in url:
            if "id=" in url:
                file_id = url.split("id=")[-1].split("&")[0]
            elif "/d/" in url:
                file_id = url.split("/d/")[1].split("/")[0]
            else:
                return url  
            return f"https://drive.google.com/uc?export=download&id={file_id}"
        return url

    df[url_col] = df[url_col].apply(convert_drive_url)
    return df

```

```python
import json
from langchain_core.documents import Document

def document_lang(df):
    docs = []
    for idx, row in df.iterrows():
        meta = json.loads(row['metadata']) if isinstance(row['metadata'], str) else row['metadata']
        # ensure meta is a dict (in case metadata is missing)
        meta = meta if isinstance(meta, dict) else {}
        meta = meta.copy()  # เพื่อไม่แก้ของเดิม
        meta['ids'] = idx   # เพิ่ม ids ตรงนี้
        doc = Document(
            page_content=row['TH_Description'],
            metadata=meta
        )
        docs.append(doc)
    return docs
```

Transform Metadata

```python
def make_metadata_chroma_safe(metadata):
    safe_meta = {}
    for k, v in metadata.items():
        if isinstance(v, (dict, list)):
            safe_meta[k] = json.dumps(v, ensure_ascii=False)
        else:
            safe_meta[k] = v
    return safe_meta

def transform_meta(docs):
    new_docs = []
    for doc in docs:
        # เช็คว่าเป็น Document (จาก LangChain) หรือ dict
        if hasattr(doc, "metadata"):
            doc.metadata = make_metadata_chroma_safe(doc.metadata)
            new_docs.append(doc)
        elif isinstance(doc, dict) and "metadata" in doc:
            doc["metadata"] = make_metadata_chroma_safe(doc["metadata"])
            new_docs.append(doc)
        else:
            print("Warning: doc is not recognized", doc)
            new_docs.append(doc)  # append original, or skip
    return new_docs
```

```python
from langchain_community.vectorstores import Chroma

def create_vecstore(embedding_function, col_name, docs, path=vec_db_path):
    
    ids = [str(doc.metadata["ids"]) for doc in docs]  # ต้องเป็น str!
    
    vector_store = Chroma(
        persist_directory=path,
        embedding_function=embedding_function,
        collection_name=col_name
    )
  
    vector_store.add_documents(docs, ids=ids)
    return vector_store
```

```python
vector_store = create_vecstore(
    embedding_function=embedding_function,
    col_name=col_name1,
    docs=shirt_docs,
    path=vec_db_path
)
```

```python
def retriver(vectorstore):
    retriever = vectorstore.as_retriever(
    search_type="similarity",
    search_kwargs={
            "k": 5,
        }
    )
    return retriever

```

Pull data from database

```python
import matplotlib.pyplot as plt
import requests
from PIL import Image
from io import BytesIO
import sqlite3

def get_img_url_from_db(db_path, table, idx, img_col="URL"):
    conn = sqlite3.connect(db_path)
    cursor = conn.cursor()
    cursor.execute(
        f"SELECT {img_col} FROM {table} WHERE id=?", (idx,)
    )
    result = cursor.fetchone()
    conn.close()
    if result:
        return result[0]
    else:
        return None

def plot_images_from_docs_db(docs, db_path, table, img_col="URL", n_max=5):
    plt.figure(figsize=(5, 3 * n_max))
    for i, doc in enumerate(docs[:n_max]):
        idx = doc.metadata['ids']
        img_url = get_img_url_from_db(db_path, table, idx, img_col=img_col)
        try:
            if not img_url:
                print(f"Not found img for idx={idx}")
                continue
            response = requests.get(img_url)
            img = Image.open(BytesIO(response.content))
            plt.subplot(n_max, 1, i+1)
            plt.imshow(img)
            plt.axis('off')
            plt.title(f"{doc.metadata.get('product_name', '')}\n(ids: {idx})")
        except Exception as e:
            print(f"Error loading image for idx={idx}: {e}")
    plt.tight_layout()
    plt.show()
```

```python
table = "pants"

plot_images_from_docs_db(docs, db_path, table, img_col="URL", n_max=5)
```
